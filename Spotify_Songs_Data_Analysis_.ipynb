The analysis focuses on two datasets, sp_tracks (containing Spotify track information) and sp_features (containing audio features for tracks).

**Popularity Analysis:**

The range of 'popularity' is determined.
The number of tracks with maximum popularity is identified.

The top 10 most popular songs (with popularity > 90) are identified and displayed with their names, popularity, and artists.

**Duration Analysis:**

A histogram of track duration (in seconds) is plotted.
The number of tracks with a duration of more than 10 minutes is calculated.
The longest song in the dataset is identified.
The top 10 longest songs are identified and displayed.

**Loudness vs Energy Analysis:**

A scatter plot is created to visualize the relationship between 'loudness' and 'energy' for a sample of the data.

**Release Date Analysis:**

The 'release_date' column is converted to datetime objects.
The distribution of release years is visualized using a histogram.

Based on the analysis conducted, several insights about the Spotify tracks dataset can be drawn:
